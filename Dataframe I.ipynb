## Rank values with the rank method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba
```

```python
# ranking by salary, highest as 1, second highest as 2
# begin by pulling salary column and replacing missing values 
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
nba
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype(int)
```

```python
nba
```

```python
nba.sort_values("Salary", ascending = False).head(11)
```
