## The groupby method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
fortune.head()
```

```python
# sums of revenues for each sector (not the best way to do)
fortune[fortune["Sector"] == "Retailing"]["Revenue"].sum()
```

```python
sectors = fortune.groupby("Sector")
sectors
```

```python
# tells us how many unique sectors there are
len(sectors)
```

```python
# how many rows fall under each sector
sectors.size()
```

```python
# pulls the first row in fortune for each sector
sectors.first()
```

```python
sectors.last()
```

## Retrieve a group with the get_group method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")

fortune.head()
```

```python
sectors.get_group("Energy")
```

## Methods on the GroupBy object

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")

fortune.head()
```

```python
sectors["Revenue"].sum()
```

```python
sectors["Employees"].sum()
```

```python
sectors["Profits"].max()
```

```python
sectors["Employees"].mean()
```

```python
sectors[["Revenue","Profits"]].mean()
```

## Grouping by multiple columns

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"])

fortune.head()
```

```python
sectors.size()
```

```python
sectors["Revenue"].sum()
```

## The agg method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")

fortune.head()
```

```python
sectors.agg({"Revenue":"sum"})
```

```python
sectors.agg({"Revenue":"sum", "Profits":"max"})
```

## Iterating through groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"])

fortune.head()
```

```python
# find the two companies in each sector with the most employees 

fortune.nlargest(2, "Employees")
```

```python
# find the two companies in each sector with the most employees 

def top_two_companies_by_employee_count(sector):
    return sector.nlargest(2, "Employees")

sectors.apply(top_two_companies_by_employee_count)
```
